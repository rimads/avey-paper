```python
import json
import os
dir = "."


def load(filepath: str):
    with open(f"{dir}/{filepath}", "r") as file:
        return json.load(file)


files = [f for f in os.listdir(dir) if os.path.isfile(os.path.join(dir, f)) and "ipynb" not in f and "experiments" not in f]
experiments = [load(f) for f in files]
# cases
```

```python
joined = [case for cases in experiments for case in cases]
joined
```

```python
with open("experiments.json","w") as file:
    json.dump(joined, file, indent=4)
```
